```python
import pickle
from scaccordion import tl as actl
import conorm
import numpy as np
```

```python
f=pickle.load(open("/home/james/sciebo/Zenodo_scACCordion/data/Distance_results/Peng_PDAC_processed_data_tmm_final.pickle","rb"))
```

```python
def norm(x):
    '''
        Defining the normalization function.
    '''
    nf = conorm.tmm(x+1e-6,trim_lfc=0,trim_mag=0)
    return (nf)
```

```python
f2 = actl.Accordion(tbls=f.tbls,weight='lr_means',normf=norm,filter=0.2)
```

```python
np.round(f2.p-f.p,)
```
