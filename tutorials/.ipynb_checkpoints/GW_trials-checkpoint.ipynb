## Load Libraries

````python
from accordion import pdac,utils
from accordion.Accordion import Accordion 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pydiffmap as dm
import pandas as pd
import networkx as nx
from scipy.linalg import  norm
import numpy as np
from scipy.sparse import diags
from scipy.sparse.linalg import eigsh
from sklearn import manifold
from sklearn.preprocessing import Normalizer
def get_Q_matrix(M):
    # taken from https://github.com/zboyd2/hitting_probabilities_metric/blob/master/HittingTimes_L3.m
    N = M.shape[0]
    e1 = np.zeros((1, N))
    e1[0][0] = 1
    A1inv = np.eye(N) - M
    A1inv[0, :] = e1
    A1inv = np.linalg.inv(A1inv)
    Q = np.zeros((N, N))
    Q[:, 0] = A1inv[:, 0].reshape(N,) / np.diagonal(A1inv)
    M = np.matmul(M, A1inv)
    detCj = ((1 + np.diagonal(M)) * (1 - M[0, 0])) + np.multiply(M[:, 0],M[0, :].transpose()).reshape(N,)
    CjInv = np.zeros((2, 2, N))
    CjInv[0, 0, :] = (1 - M[0, 0]) / detCj
    CjInv[0, 1, :] = M[:, 0].reshape(N,) / detCj # WAS WRONG INDEX HERE
    CjInv[1, 0, :] = -M[0, :].transpose().reshape(N,) / detCj
    CjInv[1, 1, :] = (1 + np.diagonal(M)) / detCj
    M1 = np.zeros((N, 2, N))
    M1[:, 0, :] = A1inv
    M1[:, 1, :] = np.tile(-A1inv[:, 0],( 1, N)).reshape(N, N)
    M2 = np.zeros((2, N, N))
    M2[0, :, :] = M.transpose()
    M2[1, :, :] = np.tile(M[0, :],(N, 1)).transpose()
    for j in range(1, N):
        # assert np.all(
        #    np.concatenate([A1inv[:, j].reshape((N, 1)), - A1inv[:, 0].reshape((N, 1))], axis=1) == M1[:, :, j])
        Ac = A1inv[:, j].reshape(N,) - np.matmul(np.matmul(M1[:, :, j], CjInv[:, :, j]), M2[:, j, j])
        Ad = np.diagonal(A1inv) - sum(
            M1[:, :, j].transpose() * np.matmul(CjInv[:, :, j], M2[:, :, j]))
        Q[:, j] = Ac / Ad
        np.fill_diagonal(Q, 0)
    return Q  # - np.diagonal(np.diagonal(Q))

def get_dhp(P, beta=0.5):
    n = P.shape[0]
    # Compute hitting probabilities
    Q = get_Q_matrix(P)
    # Find the invariant measure
    v = eigsh(P.T.astype('double'), k=1, sigma=1, which='LM',tol=1e-3,maxiter=100000,v0=np.ones(n)/n)[1].ravel()
    v1 = np.abs(v)/norm(v,ord=1)
    #v = v-np.min(v) + 1e-10
    # Construct the symmetric adjacency matrix M:
    if beta == 0.5:
        Aht = diags(v1**0.5, 0,(n, n)) @ Q @ diags(v1**-0.5, 0,(n, n))
    elif beta == 1:
        Aht = diags(v1, 0,(n, n)) @ Q 
    else:
        raise ValueError(f"Unsupported beta value: {beta}")
    Aht = (Aht + Aht.T) / 2
    np.fill_diagonal(Aht, 0)
    
    # Compute the hitting probability distance
    dhp = -np.log10(Aht,where=Aht != 0)
    np.fill_diagonal(dhp, 0)
    return(np.round(dhp,5)) 
````

````python
import ot
````

````python
gwdist ={}
for i in AaccPDAC.graphs.keys():
    gwdist[i] ={}
    for j in AaccPDAC.graphs.keys():
        if i != j:
            p1 = np.ones(len(AaccPDAC.graphs[i].nodes()))/len(AaccPDAC.graphs[i].nodes())
            q1 = np.ones(len(AaccPDAC.graphs[j].nodes()))/len(AaccPDAC.graphs[j].nodes())
            c1 = distances.ctd_dist(AaccPDAC.graphs[i])
            c2 = distances.ctd_dist(AaccPDAC.graphs[j])
            gwdist[i][j]=ot.gromov.gromov_wasserstein2(C1=c1,
                                              C2=c2,
                                              p=p1,
                                              q=q1,
                                              loss_fun='square_loss')
````

````python
gwdist_dhalf ={}
for i in AaccPDAC.graphs.keys():
    gwdist_dhalf[i] ={}
    for j in AaccPDAC.graphs.keys():
        if i != j:
            p1 = np.ones(len(AaccPDAC.graphs[i].nodes()))/len(AaccPDAC.graphs[i].nodes())
            q1 = np.ones(len(AaccPDAC.graphs[j].nodes()))/len(AaccPDAC.graphs[j].nodes())
            tmpc1 = nx.adjacency_matrix(AaccPDAC.graphs[i],weight='lr_means').todense()
            for idx in range(tmpc1.shape[0]):
                tmpc1[idx,:] = Normalizer(norm='l1').fit_transform(np.asarray(tmpc1[idx,:]))
            tmpc2 = nx.adjacency_matrix(AaccPDAC.graphs[j],weight='lr_means').todense()
            for idx in range(tmpc2.shape[0]):
                tmpc2[idx,:] = Normalizer(norm='l1').fit_transform(np.asarray(tmpc2[idx,:]))
            c1 = get_dhp(np.round(tmpc1,4))
            c2 = get_dhp(np.round(tmpc2,4))
            gwdist_dhalf[i][j]=ot.gromov.gromov_wasserstein2(C1=c1,
                                              C2=c2,
                                              p=p1,
                                              q=q1,
                                              loss_fun='square_loss')
````

````python
gwdist_done ={}
for i in AaccPDAC.graphs.keys():
    gwdist_done[i] ={}
    for j in AaccPDAC.graphs.keys():
        if i != j:
            p1 = np.ones(len(AaccPDAC.graphs[i].nodes()))/len(AaccPDAC.graphs[i].nodes())
            q1 = np.ones(len(AaccPDAC.graphs[j].nodes()))/len(AaccPDAC.graphs[j].nodes())
            tmpc1 = nx.adjacency_matrix(AaccPDAC.graphs[i],weight='lr_means').todense()
            for idx in range(tmpc1.shape[0]):
                tmpc1[idx,:] = Normalizer(norm='l1').fit_transform(np.asarray(tmpc1[idx,:]))
            tmpc2 = nx.adjacency_matrix(AaccPDAC.graphs[j],weight='lr_means').todense()
            for idx in range(tmpc2.shape[0]):
                tmpc2[idx,:] = Normalizer(norm='l1').fit_transform(np.asarray(tmpc2[idx,:]))
            c1 = get_dhp(np.round(tmpc1,4),beta=1)
            c2 = get_dhp(np.round(tmpc2,4),beta=1)
            gwdist_done[i][j]=ot.gromov.gromov_wasserstein2(C1=c1,
                                              C2=c2,
                                              p=p1,
                                              q=q1,
                                              loss_fun='square_loss')
````
